#### Import Libraries & Read in Data

```python
## standard imports 
import pandas as pd 
import numpy as np
import re
## visualizations
import matplotlib.pyplot as plt
import seaborn as sns
## preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.dummy import DummyClassifier
## modeling
from sklearn.linear_model import LogisticRegression, LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.svm import SVC, SVR
from sklearn.naive_bayes import MultinomialNB
## trees
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier, BaggingRegressor, RandomForestClassifier, ExtraTreesClassifier, RandomForestRegressor, ExtraTreesRegressor, AdaBoostRegressor, AdaBoostClassifier, GradientBoostingRegressor
## NLP
from sklearn.feature_extraction.text import CountVectorizer
## analysis
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, make_scorer, f1_score, mean_squared_error

## options
import sklearn
pd.options.display.max_rows = 4000
pd.options.display.max_columns = 100
pd.set_option('max_colwidth', 100)
```

```python
### read in data
data = pd.read_csv('../data/country.csv')
data.head()
```

```python
data['popularity'].hist()
```

```python
data.describe()
```

```python
data['is_popular'] = np.where(data['popularity'] >= 45, 1, 0) 
```

```python
data_test = data.drop(columns=['genre', 'artist_name', 'track_name', 'track_id', 'duration_ms', 'popularity']).copy()
```

```python
data_test.head()
```

```python
data['is_popular'].value_counts()
```

```python
sns.pairplot(data_test, hue='is_popular')
```

```python
### select data
X = data.drop(columns=['genre', 'artist_name', 'track_name', 'track_id', 'duration_ms', 'popularity', 'is_popular'])
y = data['is_popular']
```

```python
### TTS
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Linear Regression <a class="anchor" id="lr"></a>
<hr/>

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

print('Training score: ', lr.score(X_train, y_train))
print('Testing score: ', lr.score(X_test, y_test))
```

## Logistic Regression <a class="anchor" id="lgr"></a>
<hr/>

```python
lgr = LogisticRegression(max_iter=1000)
lgr.fit(X_train, y_train)

print('Training score: ', lgr.score(X_train, y_train))
print('Testing score: ', lgr.score(X_test, y_test))
```

```python
pipe_lgr = make_pipeline(StandardScaler(), LogisticRegression())
params_lgr = {
#     'selectkbest__k': [10],
#     'logisticregression__max_iter':[1000],
    'logisticregression__C': [0.01, 0.1, 1]
}

gs_lgr = GridSearchCV(pipe_lgr, params_lgr, n_jobs=-1)
gs_lgr.fit(X_train, y_train)

print('Training score: ', gs_lgr.score(X_train, y_train))
print('Testing score: ', gs_lgr.score(X_test, y_test))
print('Best Params', gs_lgr.best_params_)
```

## KNN Classifier <a class="anchor" id="knn"></a>
<hr/>

```python
pipe_knnC = make_pipeline(StandardScaler(), KNeighborsClassifier())

params_knnC = {
#     'selectkbest__k': [5, 10, 15],
    'kneighborsclassifier__n_neighbors': range(1, 150, 10),
}

gs_knnC = GridSearchCV(pipe_knnC, params_knnC, n_jobs=-1)
gs_knnC.fit(X_train, y_train)

print('Training score: ', gs_knnC.score(X_train, y_train))
print('Testing score: ', gs_knnC.score(X_test, y_test))
print('F1 score: ', f1_score(y, gs_knnC.predict(X)))
print('Best Params', gs_knnC.best_params_)
```
