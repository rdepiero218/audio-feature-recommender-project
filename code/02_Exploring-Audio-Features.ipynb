## Function to Pull Audio Features for Album<a class="anchor" id="write-function"></a>
<hr/>

The following function pulls audio features and popularity for a given album on Spotify. 

The first version (commented below) initially used the `album_tracks` method from the Spotipy API. It was discovered that this returns a 'simplifiied' version of the track information which did not include a track's popularity. The function was rewritten to pull track information separately so that the popularity feature could be collected and combined with the audio features for each track.

```python
def get_album_audio_features(ID):
    album_tracks_list = sp.album_tracks(ID, market='US')['items']
    
    album_tracks_URI = [album_tracks_list[i].get('uri') for i in range(len(album_tracks_list))]

    track_list = []
    
    raw_track_list = sp.tracks(album_tracks_URI, market='US')['tracks']
    
    track_list = []
    
    for i in range(len(raw_track_list)):
        track_info = [raw_track_list[i].get('name'), raw_track_list[i].get('track_number'), raw_track_list[i].get('uri'), raw_track_list[i].get('popularity')]
        track_list.append(track_info)

    track_info_df = pd.DataFrame(track_list, columns=['track_name', 'track_number', 'track_uri', 'popularity'])
    
    ### recast popularity so it's a decimal value and not out of 100?

    track_audio_features = sp.audio_features(tracks=track_info_df['track_uri'].values.tolist())

    audio_features_df = pd.DataFrame.from_dict(track_audio_features)

    drop_cols = ['type', 'id', 'uri', 'track_href', 'analysis_url']

    audio_features_df.drop(columns = drop_cols, inplace=True)

    album_df = pd.concat([track_info_df, audio_features_df], axis=1)
    
    return album_df
```

```python
def plot_audio_features(df, artist_name, album_name):
    df = df.iloc[::-1]
    ### Horizontal subplots
    df.plot.barh(
        x = 'track_name',
        y = ['valence','energy', 'danceability'],
        ylim = [0,1], 
        sharey = True,
        subplots = True, 
        layout = (1,3),
        figsize = (15,5),
        legend = False, 
        title = f'{artist_name} - {album_name}', 
        xlabel = 'Track Name')
```

### Time Duration

Duration of tracks is given in milliseconds. This isn't really a helpful unit for humans to use. I wrote a small function to convert milliseconds into a format that displays like the usual track time we're used to seeing in audio players.

```python
def convert_duration(time_ms):
    secs = int((time_ms/1000)%60)
    if secs < 10:
        secs = str('0') + str(secs)
    mins = int((time_ms/(1000*60))%60)
    mins_secs = str(mins) + ':' + secs
    return mins_secs
```

### Popularity
**From Spotify:** "The value will be an integer between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity. 
Note that the popularity value may lag actual popularity by a few days: the value is not updated in real time."

Since popularity is based on total number of plays, older music that was not on Spotify when it was current (and most likely it's most popular) is not accurately represented. If Spotify tends to recommend popular aritists and tracks to users, how are new or independent artists supposed to compete?

```python
luckier.plot.bar(x = 'track_name', y='popularity', legend=False, title = 'Luckier By Half - Popularity', xlabel='Track Name');
```

A few things to note about popularity here. The tracks are presented in order of how they appear on the album. Usually, people listen to an album from it's start, meaning tracks at the beginning of the album tend to get played more. We can see this reflected in the higher popularity ratings for the earlier tracks. (Even though it's my personal opinion that the track 'Daves Street' is one of the best tracks on this particular album).

Given this information, how is the popularity of the track 'Boston' so high? Although it's in the middle of the album it has the highest popularity. I know that this can be connected to the fact that this song was played one day on the morning news. The artist recevied a huge spike in new streams as a result which is clearly reflected in the popularity of the track here.

It's clear that popularity is not going to be a good way to necessarily rate how good a track is. However, because Spotify uses this as part of the way in which they recommend music to users, it will be an important metric to use.

## Explaining Spotify Audio Features <a class="anchor" id="explain-audio-features"></a>
<hr/>

Now we'll explore some of the audio features that Spotify provides for every track in their catalog. Nearly all of these features are obtained via direct analysis of the audio of the track. Of course these features are not necessarily perfect measures of what is happening in a given track but they are a good starting point to score and rate particular tracks.

In the future, I'll be exploring how to obtain some of these features (or perhaps some new ones of my own) by directly examining the audio analysis myself. For now, we'll just use the features that Spotify provides. A few of these are not clear as to what they score so each of the key features will be defined and explored here.

### Valence

From Spotify: "A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)." 

```python
luckier.plot.bar(x = 'track_name', y='valence', ylim=[0,1], legend=False, title = 'Luckier By Half - Valence', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Energy

From Spotify: "Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy."

```python
luckier.plot.bar(x = 'track_name', y='energy', ylim=[0,1], legend=False, title = 'Luckier By Half - Energy', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Loudness

**From Spotify:** "The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typical range between -60 and 0 db."

```python
luckier.plot.bar(x = 'track_name', y='loudness', legend=False, title = 'Luckier By Half - Loudness', xlabel='Track Name');
```

### Tempo

**From Spotify:** "The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration."

```python
luckier.plot.bar(x = 'track_name', y='tempo', legend=False, title = 'Luckier By Half - Tempo', xlabel='Track Name');
```

### Danceability

**From Spotify:** "Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable."

```python
luckier.plot.bar(x = 'track_name', y='danceability', ylim=[0,1], legend=False, title = 'Luckier By Half - Danceability', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Speechiness

**From Spotify:** "Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks."

```python
luckier.plot.bar(x = 'track_name', y='speechiness', ylim=[0,1], legend=False, title = 'Luckier By Half - Speechiness', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Acousticness

**From Spotify:** "A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic."

```python
luckier.plot.bar(x = 'track_name', y='acousticness', ylim=[0,1], legend=False, title = 'Luckier By Half - Acousticness', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Instrumentalness

**From Spotify:** "Predicts whether a track contains no vocals. Ooh and aah sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly vocal. The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0."

```python
luckier.plot.bar(x = 'track_name', y='instrumentalness', ylim=[0,1], legend=False, title = 'Luckier By Half - Instrumentalness', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

### Liveness

**From Spotify:** "Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live. "

```python
luckier.plot.bar(x = 'track_name', y='liveness', ylim=[0,1], legend=False, title = 'Luckier By Half - Liveness', xlabel='Track Name')
plt.plot([0, 13],[0.5, 0.5], 'k--');
```

#### Mode

#### Time Signature
